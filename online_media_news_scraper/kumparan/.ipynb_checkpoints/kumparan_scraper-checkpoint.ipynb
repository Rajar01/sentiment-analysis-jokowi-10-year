```python
articles = []

def extract_article(article_json):
    article = {
        'article_title' : '',
        'article_author': '',
        'article_publication_date' : '',
        'article_content' : '',
    }

    # Get the title of article
    article['article_title'] = article_json['title'].strip()

    # Get the author of article
    article['article_author'] = article_json['author']['name'].strip()

    # Get publication date of article
    article['article_publication_date'] = article_json['publishedAt'].strip()

    # Get the content of article
    article_contents = []
    artile_contents_json = json.loads(article_json['contentPublish']['document']) 
    for node in artile_contents_json['document']['nodes']:
        if node['type'] == 'paragraph':
            article_content = node['nodes'][0]['leaves'][0]['text']
            article_contents.append(article_content)

    article['article_content'] = "\n".join(article_contents).strip()
    
    return article

def store_articles_into_csv(articles):
    dataframe = pandas.DataFrame(articles)

    try:    
        if os.path.isfile('./kumparan_unprocessed_news_data.csv') == True:
            dataframe.to_csv('kumparan_unprocessed_news_data.csv', sep='\t', encoding='utf-8', index=False, header=False, mode='a')
        else:
            dataframe.to_csv('kumparan_unprocessed_news_data.csv', sep='\t', encoding='utf-8', index=False, header=True)
    
        print('Success saved article data to csv')
    except Exception as e:
        print(f'Error occured while saved article data to csv: {e}')

if __name__ == '__main__':
    with open("kumparan_api_response.json") as kumparan_api_response_json:
        article_jsons = json.loads(kumparan_api_response_json.read())
        for i in range(len(article_jsons)):
            for article_json in article_jsons[i]['data']['SearchAllContents']['edges']:
                article = extract_article(article_json['object'])
                articles.append(article)
    
        store_articles_into_csv(articles)
```
